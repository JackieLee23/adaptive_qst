```python
import numpy as np

# Perform purification using Cholesky decomposition
cholesky_factor = np.linalg.cholesky(d_matrices)

print(cholesky_factor)

# Reshape the purified states to a 2D array
purified_states = cholesky_factor.reshape(N, d*d)

# Print the purified states
# print(purified_states)

np.sum(np.abs(purified_states) ** 2, axis = 1)
```

```python
import numpy as np

# Create the purified states array

# Reshape the purified states to form a 3D array
purified_states_3d = purified_states.reshape((N, d, d))

# Compute the partial trace to obtain the density matrices
density_matrices = np.einsum('ijk,ilk->ijl', purified_states_3d, purified_states_3d.conj())

# Print the density matrices
print(density_matrices)
```
